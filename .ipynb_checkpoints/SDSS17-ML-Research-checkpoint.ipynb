# Machine Learning Research on the Stellar Classification Dataset (SDSS17)

# Блежка за проверяващият

За предишният проект https://github.com/DimovDimo/mahalanobis-prototype-classifier получих следната оценка:
В проекта има полезни части: SDSS класификацията, custom / transparent моделите и идеята, че когато физичен закон е известен, ML не трябва да се преструва, че го е преоткрил. Проблемът е, че добрата работа се губи под много тежък и преувеличен език. Бих пренаписал голяма част от текста и бих фокусирал проекта върху може би една идея: например кога класическият физически модел е по-добър, кога ML добавя стойност и как я проверяваме. Иначе много ме кефи форматирането. Бележка: Искаше ми се да останеш с SDSS dataset-a и да нямаш halting problem така от нищото :D.

Ето как беше развит предишният проект, което може би не стана ясно заради "много тежък и преувеличен език":
1. Създадох mahalanobis за клъстеризиране и видях че се справя добре със симулирани данни, но при изтински данни обикновенните ML алгоритми се справяха по-добре.
2. Показах къде ML алгоритмите имат ограничения.
3. Понеже halting problem е нерешимо ограничение още преди да се появат ML алгоритмите, го отделих в отделна чат.
4. Проведох ML изследване на SDSS17 като сравних няколко алгоритъма.
5. Проведох ML изследване на данните от CERN, но използвах нестандартни подходи който рядко се използват при изследване в ML.

Като цяло предишният проект беше:
1. Моят алгоритъм не се справя по-добре от готовите.
2. Показвам какви ограничения имат ML алгоритмите.
3. Проведох ML изследване със стандартни и нестандартни методи.

Проекта имаше "много тежък и преувеличен език" защото се опитах да звучи академично и научно точно. И това е същият език който наблюдавах в научните статии за ML, тоест изглеждаше че е написан правилно.

Ще базирам този проект на препоръките от оценката на предишният:
"фокусирал проекта върху може би една идея: например кога класическият физически модел е по-добър, кога ML добавя стойност и как я проверяваме"
"Искаше ми се да останеш с SDSS dataset-a и да нямаш halting problem така от нищото"

Ето как ще направя този проект:
1. Ще се фокусирам само върху един dataset SDSS17 "да останеш с SDSS dataset"
2. Без никакви допълнителни алгоритмични и проблемни презентации "да нямаш halting problem"
3. Ще използвам предложената идея "една идея: например кога класическият физически модел е по-добър, кога ML добавя стойност и как я проверяваме"
4. Първо започвам със стандартното зареждане на данни и изчистване на данни и EDA.
5. После ще надградя с физически и ML модели.
6. Ще се опитам да измисля начини (поне един начин) как да проверим дали физическият или ML модела е по-добър.
